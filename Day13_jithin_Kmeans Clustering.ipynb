# Exercise -2

#### Initialize 3 clusters to start k means and plot the 3 cluster points along with previous dataset

```python
df=pd.read_excel("clara.xlsx")
v1=df.V1
v2=df.V2
X=np.array(list(zip(v1,v2)))
```

```python
def generateRandomPoint(ax):
    if ax=='x':
        return random.uniform(v1.min(),v1.max())
    else:
        return random.uniform(v2.min(),v2.max())
print(generateRandomPoint('x'))
print(generateRandomPoint('y'))

k=3
c_x=[]
c_y=[]
for i in range(k):
    c_x.append(generateRandomPoint('x'))
    c_y.append(generateRandomPoint('y'))
```

```python
plt.figure(figsize=(8,5))
plt.xlabel('V1')
plt.ylabel('V2')
plt.scatter(v1,v2,s=10,color='r')
plt.scatter(c_x,c_y,s=100,marker='+',color='b')
```

```python
c=np.array(list(zip(c_x,c_y)))
cluster=np.zeros(len(X))
for j in range(len(X)):
    dist=[]
    for i in range(len(c)):
        dist.append(distance.euclidean(X[j],c[i]))
    cluster[j]=np.argmin(dist)

plt.figure(figsize=(8,5))
plt.xlabel('V1')
plt.ylabel('V2')
plt.scatter(v1,v2,s=10,c=cluster)
plt.scatter(c_x,c_y,s=100,marker='+',c='r')
```

```python
for i in range(k):
    points=[X[j] for j in range(len(X)) if cluster[j]==i]
    c[i]=np.mean(points,axis=0)            
```

```python
c_x=np.transpose(c)[0]
c_y=np.transpose(c)[1]
plt.figure(figsize=(8,5))
plt.xlabel('V1')
plt.ylabel('V2')
plt.scatter(v1,v2,s=10,c=cluster)
plt.scatter(c_x,c_y,s=100,marker='+',c='r')
```

```python
c=np.array(list(zip(c_x,c_y)))
cluster=np.zeros(len(X))
for j in range(len(X)):
    dist=[]
    for i in range(len(c)):
        dist.append(distance.euclidean(X[j],c[i]))
    cluster[j]=np.argmin(dist)

plt.figure(figsize=(8,5))
plt.scatter(v1,v2,s=10,c=cluster)
plt.scatter(c_x,c_y,s=100,marker='+',c='r')
```

```python
for i in range(k):
    points=[X[j] for j in range(len(X)) if cluster[j]==i]
    c[i]=np.mean(points,axis=0)
c_x=np.transpose(c)[0]
c_y=np.transpose(c)[1]
plt.figure(figsize=(8,5))
plt.scatter(v1,v2,s=10,c=cluster)
plt.scatter(c_x,c_y,s=100,marker='+',c='r')
```
