## DB SCAN Use Case

Generate a sample cluster data of 750 points using SKlearn and build a clustering model using DBSCAN algorithm.

```python
from sklearn.datasets import make_moons
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
X, Y = make_moons(n_samples=750, noise=0.065)
plt.scatter(x=np.transpose(X)[0],y=np.transpose(X)[1],s=10,c='g')
```

```python
X = StandardScaler().fit_transform(X)

model =DBSCAN(eps=.3, min_samples=5)
model.fit(X)
plt.scatter(x=np.transpose(X)[0],y=np.transpose(X)[1],s=10,c=model.labels_)
    
```
